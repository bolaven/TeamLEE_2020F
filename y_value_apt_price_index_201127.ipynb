```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import csv
```

```python
col = []

with open('apt_price_2006_2020.csv', newline='') as csvfile:
    
    csvreader = csv.reader(csvfile, delimiter=',')
    for row in csvreader:
        col.append(row[0])
```

```python
col
```

```python
data = pd.read_csv('apt_price_2006_2020.csv', encoding='euc-kr', header=None)
```

```python
data.head()
```

```python
# 행열 바꾸기 
data2 = data.transpose()
data2.head()
```

```python
# 컬럼 이름 설정
data2.columns=col
data2.head()
```

```python
# 현재 1행이 컬럼명으로 되어있어서 삭제
alldata = data2.iloc[1:,:]
alldata.head()
```

```python
alldata.dtypes
```

```python
alldata['지역'] = alldata['지역'].astype('string')
```

```python
# 현재 날짜를 의미하는 '지역'  컬럼에서 year, month, yymm 컬럼 추가하기
alldata['year'] = alldata['지역'].apply(lambda x : x[:4])
alldata['month'] = alldata['지역'].apply(lambda x : x[6:8])
alldata['yymm'] = alldata['year'] + alldata['month']
alldata.head()
```

```python
alldata['서울'] = alldata['서울'].astype('float64')
```

```python
# 지역별 매매지수 type 변환,  object -> float64
alldata.iloc[:,1:29] = alldata.iloc[:,1:29].astype(np.float64)
```

# 서울 ARIMA 예측
## - 원데이터 '2006.01 ~ 2020.08 /  예측 '2020.09 ~2021.12

```python
temp_seoul = temp[['date','서울']]
temp_seoul.head()
```

```python
temp_seoul.set_index('date',inplace=True)
```

```python
temp_seoul.columns=['seoul']
temp_seoul.dtypes
```

```python
temp_seoul.shape
```

```python
temp_seoul.head()
```

```python
# plt.rc('axes', unicode_minus=False)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 원데이터
temp_seoul.plot()
# 1차 차분
diff_1 = temp_seoul.diff()
diff_1.plot()
# plot_acf(diff_1)
# plot_pacf(diff_1)
# plt.show()
```

```python
# !pip install pmdarima  # pmdarima 설치
```

```python
# ARIMA  p,d,q 값 찾기
from pmdarima.arima import auto_arima

model_arima= auto_arima(temp_seoul,trace=True, error_action='ignore', start_p=1,start_q=1,max_p=3,max_q=3,suppress_warnings=True,stepwise=False,seasonal=False)

model_arima.fit(temp_seoul)
```

```python
# AIC 값이 낮을 수록 좋은 모델
# AIC 값이 가장 낮은 p,d,q (2,2,1) 을 가지고 ARIMA 모델링
from statsmodels.tsa.arima_model import ARIMA

model = ARIMA(temp_seoul, order=(2,2,1))

# ar1 , constant p-value 0.05 수준에서 유효함
model_fit = model.fit(trend = 'c', disp=0)

print(model_fit.summary())
```

```python
# 실제 데이터와 예측한 데이터 그래프
model_fit.plot_predict(dynamic=False)
plt.show()
```

```python
pr_after = pd.date_range(start = '2020-09-01',
                   	   end = None,
                       periods = 16,
                       freq = 'MS')
a = list(pr_after)
```

```python
model = ARIMA(temp_seoul, order=(2, 2, 1))  
fitted = model.fit(disp=-1)  

# Forecast
fc, se, conf = fitted.forecast(16, alpha=0.05)  # 95% conf

# Make as pandas series
fc_temp_seoul = pd.Series(fc, index=a)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(temp_seoul, label='Seoul')
plt.plot(fc_temp_seoul, label='Forecast')
plt.title('Seoul Forecast')
plt.legend(loc='upper left', fontsize=10)
plt.show()
```
